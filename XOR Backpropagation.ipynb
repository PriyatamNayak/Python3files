```python
import math

def sigmoid(net):
  return 1 / (1 + math.exp(-net*2))
x=[0,0,1,1]
z=[0,1,0,1]
t=[0,1,1,0]

w11=w21=w12=w22=w13=w23=0.5

n=0.05 #learning rate 
err=1
print("Backpropagation by Kal EL")

while abs(err)>0.04:
    for i in range(len(x)):
        y1=w11*x[i]+w21*z[i]
        f1=sigmoid(y1)
        y2=w12*x[i]+w22*z[i]
        f2=sigmoid(y2)
        y3=f1*w13+f2*w23
        f3=sigmoid(y3)
      
    #error calculation
        err=(t[i]-f3)*f3*(1-f3) 
        print(err)
        err2=err*w23*(f2*(1-f2))
        err1=err*w13*(f1*(1-f1))
       
    #weight updation
        w13=w13+n*err*f1
        w23=w23+n*err*f2
        w11=w11+n*err1*x[i]
        w12=w12+n*err2*x[i]
        w21=w21+n*err1*z[i]
        w22=w22+n*err2*z[i]
        
for i in range(len(x)):
    y1=w11*x[i]+w21*z[i]
    f1=sigmoid(y1)
    y2=w12*x[i]+w22*z[i]
    f2=sigmoid(y2)
    y3=f1*w13+f2*w23
    f3=sigmoid(y3)
    print( str(x[i]) +" "+ str(z[i])+ " = " + str(f3))
    
    
```
